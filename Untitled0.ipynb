```python
# Let's first load the contents of the uploaded files and see what data is inside.
import pandas as pd

# Loading the three files to examine the data structure.
file1_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
file2_path = '/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'
file3_path = '/content/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_31753.csv'



# Load the files into pandas DataFrames
df_population = pd.read_csv(file1_path, skiprows=4)
df_country_metadata = pd.read_csv(file2_path)
df_indicator_metadata = pd.read_csv(file3_path)

# Display the first few rows of each dataframe to understand the structure
df_population.head(), df_country_metadata.head(), df_indicator_metadata.head()
```

```python
# Select the relevant columns for the year 2023 to plot the population distribution
population_2023 = df_population[['Country Name', '2023']].dropna()

# Plotting the histogram to visualize the distribution of population in 2023
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(population_2023['2023'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Population across Countries (2023)', fontsize=14)
plt.xlabel('Total Population', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.grid(True)
plt.show()
```
